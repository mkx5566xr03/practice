```python
import pandas as pd
```

```python
apple = pd.read_csv(r"C:\Users\howger\Desktop\python_practice\vegetable_price\data\apple_revised.csv")
apple
```

```python
#原始蘋果價格展示
for i in range(1,231):
    data = pd.DataFrame()
    for row in apple:
        temp = pd.DataFrame()
        temp["date"] = apple["Date"]
        temp["price_min"] = apple["Price min"]
        temp["price_max"] = apple["Price max"]
        data = pd.DataFrame(temp)

data
```

```python
#找出data有無缺失值，顯示缺失值數量
data.isna().sum()
```

```python
#data1:找出前10個蘋果價格相差額最高，顯示時間及差額
for i in range(1,231):
    data1 = pd.DataFrame()
    for row1 in data:
        temp = pd.DataFrame()
        temp["date"] = data["date"]
        temp["price_diff"] = data.price_max - data.price_min
        data1 = pd.DataFrame(temp)

data1.sort_values("price_diff",ascending=False).head(10)
```

```python
#data2:找出蘋果價格相差額位在1.0~2.0間
for i in range(1,231):
    data2 = pd.DataFrame()
    for row1 in data:
        temp = pd.DataFrame()
        temp["date"] = data["date"]
        temp["price_diff"] = data.price_max - data.price_min
        data2 = pd.DataFrame(temp)
#觀察前140個蘋果差額最高至最低
print(data2.sort_values("price_diff",ascending=False).head(140))
print("------------------------------------")
#取第1~139位即為答案
print(data2.sort_values("price_diff",ascending=False)[1:139])
```
