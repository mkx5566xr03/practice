```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time 
```

```python
options = Options()
options.chrome_executable_path = r"C:\Users\howger\Desktop\python_practice\chromedriver.exe"
```

```python
driver = webdriver.Chrome(options=options)
driver.get("https://www.ptt.cc/bbs/Lifeismoney/index.html")
print(driver.page_source)
driver.close()
```

```python
driver = webdriver.Chrome(options=options)
driver.get("https://www.ptt.cc/bbs/Lifeismoney/index.html")
data = driver.find_elements(By.CLASS_NAME,"title")
for data_all in data:
    print(data_all.text)
print('-------------------------------')
link = driver.find_element(By.LINK_TEXT,"‹ 上頁")
link.click()
data = driver.find_elements(By.CLASS_NAME,"title")
for data_all in data:
    print(data_all.text)

driver.close()
```

```python
driver = webdriver.Chrome(options=options)
driver.get("https://www.linkedin.com/jobs/search?trk=guest_homepage-basic_guest_nav_menu_jobs&position=1&pageNum=0")
n=0
while n<3:
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
    time.sleep(5)
    n += 1
titletags = driver.find_elements(By.CLASS_NAME,"base-search-card__title")
for titletag in titletags:
    print(titletag.text)

driver.close()
```
